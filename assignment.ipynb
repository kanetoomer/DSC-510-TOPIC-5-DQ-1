```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# I'm generating my random data
np.random.seed(0)
X = np.random.rand(50)
Y = X * 2 + np.random.normal(0, 0.1, 50)

# I'm plotting that data
plt.scatter(X, Y)
plt.title('Scatter Plot of Variables X and Y')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.show()
```
